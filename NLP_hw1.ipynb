```python
import RAKE
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop = stopwords.words('russian')
```

```python
rake = RAKE.Rake(stop)
```

```python
!pip install pymorphy2
```

```python
from pymorphy2 import MorphAnalyzer
from pymorphy2.tokenizers import simple_word_tokenize
```

```python
m = MorphAnalyzer()
def normalize_text(text):
    lemmas = []
    for t in simple_word_tokenize(text):
        lemmas.append(
            m.parse(t)[0].normal_form
        )
    return ' '.join(lemmas)
```

```python
kw_lists_norm = []
for text in texts:
  kw_lists_norm.append(rake.run(normalize_text(text), maxWords=3, minFrequency=1))
```

```python
for i in kw_lists_norm:
  print(i)
```

Приведем к нормальному виду.

```python
kw_rake = []
for i in kw_lists_norm:
  x = []
  if i != []:
    for j in i:
      x.append(j[0])
  kw_rake.append(x)
print(kw_rake)
```

Теперь сделаем с помощью текстрэнка

```python
!pip3 install summa
```

```python
from summa import keywords
```

```python
kw_textrank = []
for text in texts:
  kw_textrank.append(keywords.keywords(normalize_text(text), language='russian', additional_stopwords=stop, scores=True))
```

```python
kw2_textrank = []
for i in kw_textrank:
  x = []
  for j in i:
    x.append(j[0])
  kw2_textrank.append(x)
print(kw2_textrank)
```

Текстрэнк готов, теперь генсим.

```python
from gensim.summarization import keywords as kw
```

```python
kw_gensim = []
for text in texts:
  kw_gensim.append(kw(normalize_text(text)).split('\n'))
```

```python
for i in kw_gensim:
  print(i)
```

Теперь сделаем список шаблонов по тем ключевым словам, которые я выбрала эталонами.

```python
patterns = ['NOUN+ADJ+NOUN', 'NOUN', 'ADJ+NOUN', 'VERB', 'NOUN+NOUN']
```

Отфильтруем все ими.

```python
new_gensim = []
for kwlist in kw_gensim:
  new_kws = []
  for collocation in kwlist:
    words = collocation.split(' ')
    tags = []
    for word in words:
      tags.append(str(m.parse(word)[0].tag.POS))
    template = '+'.join(tags)
    if template in patterns:
      new_kws.append(collocation)
  new_gensim.append(new_kws)
print(new_gensim)
```

```python
new_kwlist = []
for kwlist in kw2_textrank:
  new_kws = []
  for collocation in kwlist:
    words = collocation.split(' ')
    tags = []
    for word in words:
      tags.append(str(m.parse(word)[0].tag.POS))
    template = '+'.join(tags)
    if template in patterns:
      new_kws.append(collocation)
  new_kwlist.append(new_kws)
print(new_kwlist)
new_textrank = new_kwlist
```

```python
new_kwlist = []
for kwlist in kw_rake:
  new_kws = []
  for collocation in kwlist:
    words = collocation.split(' ')
    tags = []
    for word in words:
      tags.append(str(m.parse(word)[0].tag.POS))
    template = '+'.join(tags)
    if template in patterns:
      new_kws.append(collocation)
  new_kwlist.append(new_kws)
print(new_kwlist)
new_rake = new_kwlist
```

Теперь скачаем мои эталонные ключевые слова.

```python
with open('keywords.txt', encoding='utf-8') as f:
  text = f.read()
lines = text.split('\n')
print(lines)
```

```python
standard = []
for line in lines:
  s = line.split(', ')
  ns = []
  for i in s:
    ns.append(normalize_text(i))
  standard.append(ns)
print(standard)
```

Теперь можно считать метрики!

```python
import math
def precision(some_kw):
  t = []
  for i in range(len(some_kw)):
    per = 0
    for j in some_kw[i]:
      if j in standard[i]:
        per += 1
    if some_kw[i]:
      t.append(per/len(some_kw[i]))
    else:
      t.append(math.inf)
  print(t)
  return(t)
```

```python
def recall(some_kw):
  t = []
  for i in range(len(some_kw)):
    per = 0
    for j in some_kw[i]:
      if j in standard[i]:
        per += 1
    t.append(per/len(standard[i]))
  print(t)
  return(t)
```

```python
def fscore(t1, t2):
  t3 = []
  for i in range(len(t1)):
    if t1[i]+t2[i] != 0:
      f = 2*t1[i]*t2[i]/(t1[i]+t2[i])
    else:
      f = 0
    t3.append(f)
  return(t3)
```

```python
def mean(a):
  sum = 0
  num = 0
  for i in a:
    if not math.isnan(i):
      sum += i
      num += 1
  return(sum/num)
```

Выведем метрики для всего, сначала пресижн, потом средний пресижн, потом рекол, потом средний рекол, потом ф-скор, потом средний ф-скор

```python
t1 = precision(kw_rake) #рейк без шаблонов
print(mean(t1)) 
t2 = recall(kw_rake)
print(mean(t2)) 
t3 = fscore(t1, t2)
print(t3)
print(mean(t3))
```

```python
t1 = precision(new_rake) #рейк с шаблонами
print(mean(t1)) 
t2 = recall(new_rake)
print(mean(t2)) 
t3 = fscore(t1, t2)
print(t3)
print(mean(t3))
```
